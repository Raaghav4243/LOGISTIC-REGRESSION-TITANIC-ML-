```python
import pandas  as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("train.csv")

df1=pd.read_csv("test.csv")
df1.head()
```

```python
del df['Cabin']
del df1['Cabin']
```

```python
df['Age'].fillna((df['Age'].mean()),inplace=True)
df1['Age'].fillna((df1['Age'].mean()),inplace=True)
df1['Fare'].fillna((df['Fare'].mean()),inplace=True)

#df.dropna(subset="Age" , implace=True)
#df.reset_index(drop=True)
```

```python
df.info()
df1.info()
```

```python
del df['Ticket']
del df['Name']
del df1['Ticket']
del df1['Name']
```

```python
df['Survived'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
df['Sex']=encoder.fit_transform(df['Sex'])
df.head()
df1['Sex']=encoder.fit_transform(df1['Sex'])
df1.head()
```

```python
df.rename(columns={'Sex':'Male'}, inplace=True)
df1.rename(columns={'Sex':'Male'}, inplace=True)
```

```python
df1.head()
```

```python
#df['Embarked']=encoder.fit_transform(df['Embarked'])
df=pd.concat([df,pd.get_dummies(df['Embarked'], prefix='Embarked')],axis=1)
del df['Embarked']
df.head()
df1=pd.concat([df1,pd.get_dummies(df1['Embarked'], prefix='Embarked')],axis=1)
del df1['Embarked']
df1.head()
```

```python
df.corr()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()
x=df[['Age','Fare','Pclass']]
df[['Age','Fare','Pclass']]=pd.DataFrame(scaler.fit_transform(x))
df.describe()
x1=df1[['Age','Fare','Pclass']]
df1[['Age','Fare','Pclass']]=pd.DataFrame(scaler.fit_transform(x1))
df1.describe()
#df=pd.DataFrame(df)
#scaler fit transform expects a 2d array so u need to pass both array together, they will be stored at the respective places acutomatically
#also scaler.fittransform return a np array, so u need to convert it back to dataframe.
#scaling a single colum or a part of whole dataframe dont effect the correlations 
```

# DIVIDING THE LABEL AND FEATURE 

```python
X_train=df[["Pclass","Male","Age","SibSp","Parch","Fare","Embarked_C","Embarked_Q","Embarked_S"]]
Y_train=df["Survived"]
df1.head()
```
